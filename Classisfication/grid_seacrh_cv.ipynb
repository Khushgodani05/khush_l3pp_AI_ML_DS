GridSearchCv for one model with combinations of different Hyper parameters

```python
from sklearn.ensemble import RandomForestClassifier
rfc1=RandomForestClassifier()
```

```python
from sklearn.model_selection import GridSearchCV
gsc=GridSearchCV(
    estimator=rfc1,
    param_grid={
        "criterion":["entropy","gini"],
        "max_depth":[5,7,10,15,25],
        "n_estimators":[100,200,500]
    },
    cv=5,
    verbose=3,
    return_train_score=True,
    )
```

```python
gsc.fit(df.iloc[:,0:-1],df.iloc[:,-1:].values.ravel())
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
rfc=RandomForestClassifier()
log_reg=LogisticRegression()
svc=SVC()
naive_bayes=GaussianNB()
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
df.iloc[:,0:-1]=sc.fit_transform(df.iloc[:,0:-1])
df.head()
```

```python
models={
    "DecisionTreeClassifier":dtc,
    "RandomForestClassifier":rfc,
    "LogisticRegression":log_reg,
    "Support vector classifier":svc,
    "naive_bayes":naive_bayes,
}
params={
    "DecisionTreeClassifier":{
        "criterion":["entropy","gini"],
        "max_depth":[5,10,15],
    },
    "RandomForestClassifier":{
        "n_estimators":[100,200,500],
        "criterion":["entropy","gini"],
        "max_depth":[5,15,30],
    },
    "LogisticRegression":{
        "max_iter":[100,200,500],
    },
    "Support vector classifier":{
        "kernel":["linear","poly","rbf"],
        "C":[5,7,15],
    },
    "naive_bayes":{}
}
```

```python
GridSearchCV_score=pd.DataFrame(columns=["model","best_params","best_score"])
i=0
for names in models:
    print("Model :",names)
    clf=GridSearchCV(estimator=models[names],param_grid=params[names],cv=5,return_train_score=True)
    clf.fit(df.iloc[:,0:-1],df.iloc[:,-1:].values.ravel())
    GridSearchCV_score.loc[i]=[names,clf.best_params_,clf.best_score_]
    i=i+1
GridSearchCV_score
```
