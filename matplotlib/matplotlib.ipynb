```python
x=np.linspace(0,2,100)
fig,ax=plt.subplots()
ax.plot(x,x,c="red",label="Linear")
ax.plot(x,x**2,c="blue",label="Quadratic")
ax.plot(x,x**3,c="green",label="Cubic")
ax.set_xlabel("X label")
ax.set_ylabel("Y label")
ax.set_title("Sample graph")
plt.legend()
```

pyplot-style

```python
x=np.linspace(0,2,100)
plt.plot(x,x,c="red",label="Linear")
plt.plot(x,x**2,c="blue",label="Quadratic")
plt.plot(x,x**3,c="green",label="Cubic")
plt.plot(x,x**4,label="Quadraple")
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Simple graph")
plt.legend()
```

```python
x=np.linspace(0,3,100)
plt.scatter(x,x,c="red",label="Linear")
plt.scatter(x,x**2,c="blue",label="Quadratic")
plt.scatter(x,x**3,c="green",label="Cubic",marker="+")
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Simple graph")
plt.legend()
```
