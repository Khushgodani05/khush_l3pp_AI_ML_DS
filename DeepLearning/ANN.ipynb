# Artificial Neural Networks

Artificial neural networks (ANNs) are a type of machine learning algorithm that are inspired by the human brain. They are made up of interconnected nodes, called neurons, that are organized into layers. Data flows through these neurons, adjusting the weights of connections to learn patterns and make predictions.

A neuron is a unit of computation in a neural network. It takes in a set of inputs, performs a calculation, and outputs a single value. The inputs to a neuron are typically the outputs of other neurons. The calculation performed by a neuron is called an activation function. The activation function determines how the neuron responds to its inputs.

In brief, **a neural network is a system of interconnected neurons that can learn to perform complex tasks by adjusting the weights of its connections. A neuron is a unit of computation in a neural network that takes in a set of inputs and outputs a single value**.

## **Artificial Neural Network Layers**

Artificial Neural network is typically organized in layers. Layers are being made up of many interconnected ‘nodes’ which contain an **‘activation function’**. A neural network may contain the following 3 layers:

### **a. Input layer**

The purpose of the input layer is to receive as input the values of the explanatory attributes for each observation. Usually, the number of input nodes in an input layer is equal to the number of explanatory variables. ‘input layer’ presents the patterns to the network, which communicates to one or more ‘hidden layers’.

The nodes of the input layer are passive, meaning they do not change the data. They receive a single value on their input and duplicate the value to their many outputs. From the input layer, it duplicates each value and sent to all the hidden nodes.

### **b. Hidden layer**

The Hidden layers apply given transformations to the input values inside the network. In this, incoming arcs that go from other hidden nodes or from input nodes connected to each node. It connects with outgoing arcs to output nodes or to other hidden nodes. **In hidden layer, the actual processing is done via a system of weighted ‘connections’.** There may be one or more hidden layers. The values entering a hidden node multiplied by weights, a set of predetermined numbers stored in the program. The weighted inputs are then added to produce a single number.

### **c. Output layer**

The hidden layers then link to an ‘output layer‘. Output layer receives connections from hidden layers or from input layer. It returns an output value that corresponds to the prediction of the response variable. In classification problems, there is usually only one output node. The active nodes of the output layer combine and change the data to produce the output values.

The ability of the neural network to provide useful data manipulation lies in the proper selection of the weights. This is different from conventional information processing.

ANNs are used for a wide variety of tasks, including image recognition, speech recognition, natural language processing, and medical diagnosis. They are able to learn complex patterns from data and make accurate predictions, even in the presence of noise or incomplete data.

## **What is the activation function?**

Activation function defines the output of input or set of inputs or in other terms defines node of the output of node that is given in inputs. T**hey basically decide to deactivate neurons or activate them to get the desired output.** It also performs a nonlinear transformation on the input to get better results on a complex neural network.

Activation function also helps to normalize the output of any input in the range between **1 to -1**. Activation function must be efficient and it should reduce the computation time because the neural network sometimes trained on millions of data points.

Activation function basically decides in any neural network that given input or receiving information is relevant or it is irrelevant. Let's take an example to understand better what is a neuron and how activation function bounds the output value to some limit.

The neuron is basically is a weighted average of input, then this sum is passed through an activation function to get an output.

### **Y = ∑ (weights\*input + bias)**

Here Y can be anything for a neuron between range -infinity to +infinity. So, we have to bound our output to get the desired prediction or generalized results.

### **Y = Activation function(∑ (weights\*input + bias))**

So, we pass that neuron to activation function to bound output values.

```python
data=torch.tensor(x,dtype=torch.float32)
data,data.dtype
```

```python
from torch.utils.data import Dataset,DataLoader
```

```python
class Data(Dataset):
    def __init__(self,x,y):
        super().__init__()
        self.x=torch.tensor(x,dtype=torch.float32)
        self.y=torch.from_numpy(y.astype(np.float32))
    
    def __getitem__(self,index):
        return self.x[index], self.y[index]
    
    def __len__(self):
        return self.x.shape[0]
```

```python
Data(x_train,y_train)
```

```python
Data(x_train,y_train)[800]
```

```python
train_data=Data(x_train,y_train)
test_data=Data(x_test,y_test)
```

```python
train_dataloader=DataLoader(
    dataset=train_data,
    batch_size=64,
    shuffle=True,
    drop_last=False,
)

test_dataloader=DataLoader(
    dataset=test_data,
    drop_last=False,
    shuffle=False,
    batch_size=64
)
```

```python
for len,(x,y) in enumerate(train_dataloader):
    print(len, x,y)
```

```python
for lex,(x,y) in enumerate(test_dataloader):
    print(len,x,y)
```

```python
from torch import nn
from torch import optim

input_dim=2
hidden_dim=10
output_dim=1

class NeuralNetwork(nn.Module):
    def __init__(self,input_dim,output_dim,hidden_dim):
        super().__init__()
        self.layer_1=nn.Linear(
            in_features=input_dim,
            out_features=hidden_dim,
        )
        self.layer_2=nn.Linear(
            in_features=hidden_dim,
            out_features=output_dim
        )
    
    def forward(self,x):
        # print(x.shape)
        x=self.layer_1(x)
        # print(x.shape)
        x=nn.functional.relu(x)
        # print(x.shape)
        x=self.layer_2(x)
        # print(x.shape)
        x=nn.functional.sigmoid(x)
        # print(x.shape)
        return x
    
model=NeuralNetwork(input_dim,output_dim,hidden_dim)
print(model)
```

```python
for parameter in model.parameters():
    print(parameter)
```

```python
learning_rate=0.1
loss_fn=nn.BCELoss()
optmizer=optim.SGD(model.parameters(),lr=learning_rate)
```

```python
device=torch.device("cpu")
epoch=100
print(device)
model2=model.to(device)
model2.train()
avg_loss=[]
for epochs in range(epoch):
    loss_value=[]
    for x,y in train_dataloader:
        x=x.to(device)
        y=y.to(device)
        optmizer.zero_grad()
        # forward + backward + optimize
        pred=model2(x)
        loss=loss_fn(pred,y.unsqueeze(-1))
        loss_value.append(loss.item())
        loss.backward()  #computes the gradient of parameters
        optmizer.step() #updates the weights    
    avg_loss.append(np.mean(loss_value))
    print(f"Epoch: {epochs} | Loss: {np.mean(loss_value)}")
    avg_loss.append(np.mean(loss_value))

    
```

```python
with torch.inference_mode():
    total,correct=0,0
    for x,y in test_dataloader:
        x=x.to(device)
        y=y.to(device)
        output=model2(x)
        output=output.detach().cpu()
        predicted=np.where(output>0.5 ,1,0)
        predicted=np.squeeze(predicted,-1)
        total+=y.shape[0]
        correct+=((predicted==y).sum())
print( f"Accuracy of the network on the {total} test instances: {100 * correct / total}%")        
        
```
